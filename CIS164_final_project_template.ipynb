# Data Cleaning

Explain what needed to be fixed before analysis.

Possible tasks:
- Handling missing values
- Converting column types
- Filtering rows
- Creating new columns

Example explanation:
- Convert sleep hours to numeric
- Remove rows with missing depression data


```python
df["sleep_hours"] = pd.to_numeric(df["sleep_hours"], errors="coerce")
df_clean = df.dropna(subset=["sleep_hours", "depression_status"])
```

# Visualizations

Each visualization should follow this pattern:

Example explanation:
- This boxplot compares sleep hours across age groups


```python
sns.boxplot(x="age_group", y="sleep_hours", data=df_clean)
plt.title("Sleep Duration by Age Group")
plt.show()
```

**Interpretation Example**

- Younger age groups appear to have slightly lower sleep duration
- Older groups show less variation
- Differences appear moderate but require further analysis
